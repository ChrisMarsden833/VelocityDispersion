```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MD = 10
RD = 5

def sig(R):
    return MD/(2.*np.pi*RD**2) * np.exp(-R/RD)

def Mass(R):
    return MD/(RD**2) * ( -RD * np.exp(-R/RD) * (RD + R) + RD**2)
```

```python
R_range, step = np.linspace(0, 10*RD, 100, retstep=True)

sig_sample = sig(R_range)

integral = np.cumsum(sig_sample*step*2.*np.pi*R_range)

plt.figure(dpi = 140)
plt.plot(R_range, sig(R_range), label = "sig")
plt.plot(R_range, integral, label = "Numeric")
plt.plot(R_range, Mass(R_range), ':', label = "Analytic")
plt.legend()
plt.loglog()
plt.ylim([-1, MD+3])
plt.show()
```
